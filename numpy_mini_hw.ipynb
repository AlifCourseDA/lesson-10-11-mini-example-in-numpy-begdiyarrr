## Task 2. Power of a matrix

Write a function that would take in a matrix and a non-negative integer,
check for that matrix is square, and return the power of the matrix computed through
successive multiplication.

```python
import numpy as np

def matrix_power(mtrx,n=1):
    var = np.array(mtrx)
    for _ in range(1, n):
        var = var.dot(mtrx)
    return var
A = np.array([[1,2,0],[0,2,3],[1,1,5]])
print(matrix_power(A))
```

## Task 2. Autoregressive model in matrix form

Consider the AR(1) model

$$
y_t = a y_{t-1} + \varepsilon, \; \varepsilon \sim N(0, 1).
$$

We can represent it in the form

$$
Ay = \varepsilon \quad \quad \varepsilon \sim N(0, 1)
$$

where $ A $ is

$$
A = \begin{bmatrix} 1  & 0 & \cdots & 0 & 0  \cr
                      -a  & 1 & \cdots & 0 & 0 \cr
                      \vdots & \vdots & \cdots & \vdots & \vdots \cr
                      \vdots & \vdots & \cdots & 1 & 0 \cr
                      0 & 0 & \cdots & -a & 1 \end{bmatrix}
$$

and $ y $ and $ \varepsilon $ are $ (T x 1) $ vectors

Generate an AR(1) series with $ T=500 $ and $ \alpha = 0.9 $
using matrix algebra, and make a plot of $ y_t $.

Hint: use NumPy.eye() with additional arguments.

```python
import numpy as np

T = 500
α = 0.9
ɛ = np.random.randn(T)

import numpy as np
import matplotlib.pyplot as plt

def generate_autoregressive_data(α, T, initial_value=0, ɛ = 0.1):
    y = np.zeros(T)
    y[0] = initial_value
    for t in range(1, T):
        y[t] = α * y[t-1] + np.random.normal(0, ɛ)
    return y

def visualize_autoregressive_data(y):
    plt.plot(y, label='Autoregressive Time Series')
    plt.title('Autoregressive Model Simulation')
    plt.xlabel('Time Step')
    plt.ylabel('Value')
    plt.legend()
    plt.show()

time_series_data = generate_autoregressive_data(α, T)
visualize_autoregressive_data(time_series_data)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_autoregressive_data(a, T, initial_value=0, noise_std=0.1):
    """
    Generate data from an autoregressive model.

    Parameters:
    - a: Autoregressive coefficient
    - num_steps: Number of time steps
    - initial_value: Initial value of the time series
    - noise_std: Standard deviation of the random error term

    Returns:
    - y: Generated time series data
    """
    y = np.zeros(T)
    y[0] = initial_value
    for t in range(1, T):
        y[t] = a * y[t-1] + np.random.normal(0, noise_std)
    return y

def visualize_autoregressive_data(y):
    """
    Visualize the generated autoregressive time series data.

    Parameters:
    - y: Time series data
    """
    plt.plot(y, label='Autoregressive Time Series')
    plt.title('Autoregressive Model Simulation')
    plt.xlabel('Time Step')
    plt.ylabel('Value')
    plt.legend()
    plt.show()

# Parameters
a = 0.9  # Autoregressive coefficient
T = 500  # Number of time steps

# Generate autoregressive data
time_series_data = generate_autoregressive_data(a, T)

# Visualize the results
visualize_autoregressive_data(time_series_data)
```
